```python
import pandas as pd 
from datetime import datetime 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data=pd.read_csv("TSLA.csv",date_parser='Date')
```

```python
data.head(5)
```

```python
data.plot()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Plot the high column as a line graph
plt.plot(data['High'])

# Customize the plot for readability
plt.title('Stock High Prices Over Dates')
plt.xlabel('Date')
plt.ylabel('High Price')
```

```python
data['High'].plot(xlim=["2020-01-01","2021-09-01"],ylim=[0,900],figsize=(12,4),c='red')
```

```python
index=data.loc['2020-01-01':'2021-09-01'].index
open=data.loc['2020-01-01':'2021-09-01']['Open']
```

```python
fig,sub=plt.subplots()
sub.plot(index,open)
fig.autofmt_xdate()
plt.tight_layout()
plt.xlabel("Date")
plt.ylabel("Open")
plt.show()
```

```python
data.resample(rule="A").min()["Open"].plot()
```

```python
data.resample(rule="A").max()["Open"].plot()
```

```python
data.resample(rule="QS").max()["Open"].plot()
```

```python
data.resample(rule="M").max()["Open"].plot(kind='bar',figsize=(35,6))
```
